```python
import numpy as np

def DA_AD_product(D, A):

    DA = np.zeros((D.shape[0], A.shape[1])) 

    # Fast D @ A
    for i in range(D.shape[0]):
        for j in range(A.shape[1]):
            DA[i, j] = D[i, i] * A[i, j]

    AD = np.zeros((A.shape[0], D.shape[1]))
    
    # Fast A @ D
    for i in range(A.shape[0]):
        for j in range(D.shape[1]):
            AD[i, j] = A[i, j] * D[j, j]

    res = {
        'DA': DA,
        'AD': AD,
    }
    return res
```
